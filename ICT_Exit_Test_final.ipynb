```python
# Import the necessary library
from sklearn.preprocessing import OneHotEncoder

# Initialize the encoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Encode the categorical columns
encoded_cols = encoder.fit_transform(data[['gender', 'strong_foot', 'dedication_level', 'coaching']])

# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names(['gender', 'strong_foot', 'dedication_level', 'coaching']))

# Concatenate the encoded DataFrame with the original data
data = pd.concat([data, encoded_df], axis=1)

# Drop the original categorical columns
data.drop(['gender', 'strong_foot', 'dedication_level', 'coaching'], axis=1, inplace=True)
```

```python
import numpy as np
import pandas as pd

# Define a function to detect outliers using IQR
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Create a list to store outlier dataframes
outliers_list = []

# Iterate through numeric columns and detect outliers
numeric_columns = data.select_dtypes(include=np.number).columns
for column in numeric_columns:
    outliers = detect_outliers(data, column)
    outliers_list.append(outliers)

# Concatenate all outlier dataframes into one
outliers_df = pd.concat(outliers_list)

# Remove duplicate rows if any
outliers_df = outliers_df.drop_duplicates()

```

```python
def winsorize(data, column):
    p5 = data[column].quantile(0.05)
    p95 = data[column].quantile(0.95)
    data[column] = np.where(data[column] < p5, p5, np.where(data[column] > p95, p95, data[column]))

for column in numeric_columns:
    winsorize(data, column)
```
